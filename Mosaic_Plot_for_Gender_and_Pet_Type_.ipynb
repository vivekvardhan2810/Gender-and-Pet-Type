**Mosaic plot using Statsmodels**

```python
import pandas as pd
from statsmodels.graphics.mosaicplot import mosaic

# Create a DataFrame from your dataset
data = pd.DataFrame({
    'Gender': ['Male', 'Male', 'Male', 'Female', 'Female', 'Female'],
    'Pet Type': ['Dog', 'Cat', 'Fish', 'Dog', 'Cat', 'Fish'],
    'Count': [30, 15, 10, 25, 20, 5]
})

# Create a mosaic plot
mosaic(data, ['Gender', 'Pet Type'], title='Mosaic Plot for Gender and Pet Type')
```

**Mosaic plot by using matplotlib**

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for the mosaic plot
observed_counts = np.array([[30, 15, 10], [25, 20, 5]])

# Labels for genders and pet types
genders = ['Male', 'Female']
pet_types = ['Dog', 'Cat', 'Fish']

# Create the mosaic plot manually
fig, ax = plt.subplots(figsize=(8, 6))

for i, gender in enumerate(genders):
    bottom = 0
    for j, pet_type in enumerate(pet_types):
        observed_count = observed_counts[i, j]
        ax.bar(i, observed_count, bottom=bottom, label=pet_type)
        bottom += observed_count

# Add labels and legend
ax.set_xticks(range(len(genders)))
ax.set_xticklabels(genders)
ax.set_xlabel('Gender')
ax.set_ylabel('Count')
ax.set_title('Mosaic Plot for Gender and Pet Type')
ax.legend(title='Pet Type', loc='upper right')

plt.show()
```
